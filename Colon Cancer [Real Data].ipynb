```python
dict_outliersT = dict()
dict_names = dict()
dict_strings = dict()
outliersT=[]

for sample in diff_PU_t.index:
    for intron in diff_PU_t:
        if intron not in dict_outliersT:
            dict_outliersT[intron] = [0]
            dict_names[intron] = []
            dict_strings[intron] = [""]
        if diff_PU_t.get_value(sample, intron) > (1.5* (df_pu_t.get_value(intron, "IQR"))):
            outliersT.append((sample, intron, df_pu_t.get_value(intron, 'mean')))
            dict_outliersT[intron][0] += 1
            dict_names[intron].append(sample)
            dict_strings[intron][0] += "," + sample
```
